```python
import pandas as pd

# Reload all files
ratings = pd.read_csv("C:/Users/yuvat/AI Capstone/Part 2/tourism_rating.csv")
user = pd.read_csv("C:/Users/yuvat/AI Capstone/Part 2/user.csv")
places = pd.read_excel("C:/Users/yuvat/AI Capstone/Part 2/tourism_with_id.xlsx")

# Clean column names
ratings.columns = ratings.columns.str.strip()
user.columns = user.columns.str.strip()
places.columns = places.columns.str.strip()
places = places.loc[:, ~places.columns.str.contains('^Unnamed')]  # Remove unnamed

# Check if expected columns exist
print("\nRATINGS columns:", ratings.columns)
print("USER columns:", user.columns)
print("PLACES columns:", places.columns)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Make sure IDs are strings
ratings['User_Id'] = ratings['User_Id'].astype(str)
user['User_Id'] = user['User_Id'].astype(str)
ratings['Place_Id'] = ratings['Place_Id'].astype(str)
places['Place_Id'] = places['Place_Id'].astype(str)

# Merge the data
combined = ratings.merge(user, on='User_Id').merge(places, on='Place_Id')
print("\nMerged Shape:", combined.shape)
print("\nCombined Columns:\n", combined.columns)

# Fix column name if there's a City_x or City_y
if 'City' not in combined.columns:
    if 'City_y' in combined.columns:
        combined.rename(columns={'City_y': 'City'}, inplace=True)
    elif 'City_x' in combined.columns:
        combined.rename(columns={'City_x': 'City'}, inplace=True)

# Drop unnecessary columns
combined = combined.drop(columns=[col for col in combined.columns if col.startswith('Unnamed')], errors='ignore')

# Most loved places
liked_spots = combined.groupby('Place_Name')['Place_Ratings'].mean().sort_values(ascending=False).head(10)
print("\nMost Loved Spots:\n", liked_spots)

# Most loved cities
city_avg = combined.groupby('City')['Place_Ratings'].mean().sort_values(ascending=False).head(5)
print("\nCities with Most Loved Spots:\n", city_avg)

# Most loved categories
liked_categories = combined.groupby('Category')['Place_Ratings'].mean().sort_values(ascending=False)
print("\nTop Categories by Rating:\n", liked_categories)

# Create pivot table for recommendation
pivot = combined.pivot_table(index='User_Id', columns='Place_Name', values='Place_Ratings')
pivot.fillna(0, inplace=True)

similarity = cosine_similarity(pivot.T)
sim_df = pd.DataFrame(similarity, index=pivot.columns, columns=pivot.columns)

# Recommend similar places
def recommend(place, n=5):
    if place not in sim_df.columns:
        print("Place not found.")
        return
    print(f"\nRecommendations for: {place}\n")
    recs = sim_df[place].sort_values(ascending=False)[1:n+1]
    for i, score in recs.items():
        print(f"{i} (Similarity: {score:.2f})")
```

```python
recommend("Dunia Fantasi")
```
